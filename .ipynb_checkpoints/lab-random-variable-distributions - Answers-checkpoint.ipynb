```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('numerical.csv') 
data.head()
```

# Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.


```python
data.hist(bins=15, figsize=(15, 6), layout=(2, 4));

    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,10))
for i, column in enumerate(data.columns, 1):
    plt.subplot(3,4,i)
    sns.histplot(data[column])
```

# Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.


```python
def log_transform_clean_(x):
    if np.isfinite(x) and x!=0: 
        return np.log(x)
    else:
        return np.NAN
```

```python
for i in data.columns:
    data[i]=list(map(log_transform_clean_,data[i]))
```

```python
#after log transformation:
sns.set(rc={'figure.figsize':(2,1)})
for i in data.columns:
    sns.distplot(data[i])
    plt.show()
```

```python
# We can see that doing the transformation, we can find some normal distributions and the rest are also more normilize.
```
